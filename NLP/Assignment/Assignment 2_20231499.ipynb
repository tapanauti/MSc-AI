## Task 2 (10 Marks)
Write a function that takes a document and returns a list of sentences with part-of-speech tags.

Please keep in mind that the expected output is a list within a list as shown below.


Hint: For this task you need to perform three steps:
1. Sentence Segmentation
1. Word Tokenization
1. Part-of-Speech Tagging

```python

def ie_preprocess(document):
  tag_sentence = []
  list_of_sentences = sent_tokenize(document)   # Sentence Tokenising
  for line in list_of_sentences:
    tag_sentence.append(pos_tag(word_tokenize(line)))   #pos tagging
  return tag_sentence
  
```

## Task 3 (20 Marks)
Write a function that takes a list of tokens with POS tags for a sentence and returns a list of named entities (NE). 

Hint: Use `binary = True` while calling NE chunk function

```python
def find_named_entities(sent):
  named_entities = []
  a= []
  #extracting named entities
  a.append(ne_chunk(sent,binary=True))
  for i in a:
    for b in i:
      if hasattr(b,'label'): #checking words for named entities
        named_entities.append(' '.join(c[0] for c in b))
  
  return named_entities
```

## Task 4 (5 Marks)

Implement the `find_all_named_entities` function below to find **all** NEs in a given document.

Hint: Use `find_named_entities` implemented above for this task.

```python
def find_all_named_entities(doc):
  named_entity = []
  tagged_sentences = ie_preprocess(doc)
  for i in range(len(tagged_sentences)):
    named_entity.extend(find_named_entities(tagged_sentences[i]))  
  return named_entity  
find_all_named_entities(docs[0])

# for all unique set uncomment the below code
# res = set(find_all_named_entities(first_doc))
# print(res)
```

## Task 6 (40 Marks)

Write functions to extract the name of the player, country of origin and date of birth as well as the following relations: team(s) of the player and position(s) of the player.

Hint: Use the `re.compile()` function to create the extraction patterns.

Reference: https://docs.python.org/3/howto/regex.html

```python
import difflib
def name_of_the_player(doc):
  final = ie_preprocess(doc)
  ne = ne_chunk(final[0])
  name = []
  for i in ne.subtrees(filter=lambda t: t.label() == 'PERSON'): # if the named enitiy has label person return the named entity
    for leaf in i.leaves():
      name.append(leaf[0])

  return ' '.join(name)

def country_of_origin(doc):
  sentence = sent_tokenize(doc)
  final = ie_preprocess(doc)
  ne = ne_chunk(final[0])
  nm=[]
  country = []
  names = [ 'Argentina','Spain','Swedan' , 'Portugal','England', 'Brazil', 'England', 'Wales', 'Germany' ]      
  for i in ne.subtrees(filter=lambda t: t.label() == 'GPE'):
    for leaf in i.leaves():
      country.append(leaf[0]) #  store the location named entity in the array
  match = re.compile(r'((?:\w+\W+))(professional)') # regex for extracting the name
  fmatch = match.findall(sentence[0])
  if fmatch is None:
    nm =  [difflib.get_close_matches(i,names)for i in set(country)] #  if match not found check in named entity
  if len(fmatch) !=0:   
    country_final = difflib.get_close_matches(fmatch[0][0],country) # the returned word might be like portugese of brazilian but we need country so match this with country array and get the country name.
    nm = [difflib.get_close_matches(i,names)for i in set(country_final)]
  if len(nm)!=0:
    print(nm[0][0]) # print the country name if found any
  else:
    match = re.search(r'((?:\w+\W+))(national)', sentence[0]) # match this regex to extract the name
    if match is None:
      match = re.search(r'((?:\w+\W+))(national)', sentence[1]) # if match not found in first sentence search in second
    cntry = [difflib.get_close_matches(match.group().split(' ')[0],names)] # we may get similar country name but nt excat like welsch or german so  match this with country array and get final country name.
    print(cntry[0][0])


def date_of_birth(doc):
  sentence = sent_tokenize(doc)
  born = re.search(r'born ((\S+\s+){,3})', sentence[0])  # regex for date of birth
  date = born.group()
  date = re.sub(r'([^\s\w]|_)+', '',date[5:]) # date may come ending with attached bracket or full stop, regex to remove that
  print(date)
  

def team_of_the_player(doc):
  final = []
  sentence = sent_tokenize(doc)
  team = re.compile(r'club ((?:\S+\s+){0,2})') # club search 
  team1 = re.search(r'((?:\w+\W+))(national team)',sentence[0]) # regex for national team
  team_1 = re.search(r'((?:\w+\W+))(national team)',sentence[1])
  for i,j in enumerate(sentence):
    if i==2:
      break
    tms = team.findall(j)
    if len(tms) !=0:
      for i in tms:
         res = find_all_named_entities(i)  # the regex might return random values which  are not named entities , this filters the named entities. 
         if len(res) != 0:
           final.extend(res)
  if team1 is not None:
    final.append(team1.group())
  elif team_1 is not None:
    final.append(team_1.group())
  print(final)

def position_of_the_player(doc):
  final = []
  sentence = sent_tokenize(doc)
  position = ['forward', 'attacking midfielder','striker','winger','central midfielder']
  for x in position:
    p = re.compile(r"\b{}\b".format(x),re.I) # for each position check if that is present in the corpus return that if present.
    for i in sentence:
      if p is not None:
        pos = p.findall(i)
        if len(pos) !=0:
          final.extend(pos)
  print(final[0])
          
        
 

  
```

## Task 6 (10 Marks)
Identify one other relation (besides team and player) and write a function to extract it.

```python
def team_league(doc):
  final = []
  sentence = sent_tokenize(doc)
  position = ['UEFA Champions League',' Premier League',' FIFA Club World Cup','FIFA World cup']
  for x in position:
    p = re.compile(r"\b{}\b".format(x),re.I) # for each indidvisual check the league if present retun it.
    for i in sentence:
      if p is not None:
        pos = p.findall(i)
        if len(pos) !=0:
          final.extend(pos)
  if len(final) == 0:
      print("No data available")  # if no league information is available return no data.
  else:
      print(list(set(final)))
          
```
